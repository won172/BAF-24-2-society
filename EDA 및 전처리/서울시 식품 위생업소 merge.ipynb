업종명 기준으로 휴게음식점 분리
- 휴게음식점은 음주행위가 허용되지 않는 곳으로 주로 다류(茶類), 아이스크림류 등을 조리·판매하거나 패스트푸드점, 분식점 형태의 영업

```python
seoul['업종명'].unique()
```

```python
seoul_filtered = seoul[seoul['업종명'] == '휴게음식점']
# seoul_filtered.head()
```

- 휴게음식점 중 프랜차이즈 카페만 추출
- 카페 리스트 : '메머드', '매머드', '컴포즈', '메가', '빽다방', '더벤티', '이디야', '스타벅스', '커피빈', '할리스', '투썸플레이스','커피빈','더커피빈','카페베네','탐앤탐스','풀바셋','파스쿠찌'

```python
seoul_filtered['업소명'].str.contains('바나', regex=True, case=False)
```

```python
# '메머드', '컴포즈', '메가' 등의 프렌차이즈 카페명을 포함한 행 필터링
franchises = ['메머드', '매머드', '컴포즈', '메가', '빽다방', '더벤티', '이디야', '스타벅스', '할리스', '투썸플레이스','커피빈','더커피빈','탐앤탐스','디저트39','디저트 39','셀렉토','커피베이','바나프레소']

pattern = '|'.join(franchises)
seoul_filtered_cafe = seoul_filtered[seoul_filtered['업소명'].str.contains(pattern, regex=True, case=False)]

# 결과 출력
seoul_filtered_cafe.info() ##6393
```

```python
file_path = '/content/drive/MyDrive/baf_society/seoul_franchises_cafe0928.csv'
seoul_filtered_cafe.to_csv(file_path, index=False)
```
