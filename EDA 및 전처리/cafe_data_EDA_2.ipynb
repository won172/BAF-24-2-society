```python
# 행정동을 추출하여 추가할 리스트 생성
administrative_dongs = []
# 동 정보 추출 함수

# 괄호 안에서 '~동'만 추출하는 함수 정의 (문자열 여부 확인 추가)
def extract_dong(address):
    if isinstance(address, str):  # 문자열인지 확인
        # 괄호 안에서 첫 번째 '동' 매칭
        match = re.search(r'\(([^,]*동)', address)
        return match.group(1).strip() if match else None
    return None  # 문자열이 아닌 경우 None 반환

# 모든 행에 대해 동 정보 추출 및 새로운 열에 저장
df_filtered['추출된 동'] = df_filtered['소재지도로명'].apply(extract_dong)

# 결측치가 있는 '행정동' 열에 대해 '추출된 동' 정보로 채우기
df_filtered.loc[df_filtered['행정동명'].isna(), '행정동명'] = df_filtered['추출된 동']
```

```python
# 날짜 형식이 올바르지 않을 수 있으므로 날짜 형식으로 변환
df_filtered_dong['영업자시작일'] = pd.to_datetime(df_filtered_dong['영업자시작일'], errors='coerce')
df_filtered_dong['폐업일자'] = pd.to_datetime(df_filtered_dong['폐업일자'], errors='coerce')

# 총영업일 계산
기준일 = pd.to_datetime('2023-09-28')
df_filtered_dong['총영업일'] = df_filtered_dong.apply(
    lambda row: (row['폐업일자'] - row['영업자시작일']).days if pd.notna(row['폐업일자']) else (기준일 - row['영업자시작일']).days,
    axis=1
)
```

```python
# 월 추출
df_filtered_dong['월'] = df_filtered_dong['영업자시작일'].dt.month

# 계절 정의
def define_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    elif month in [12, 1, 2]:
        return '겨울'
    return None

df_filtered_dong['계절'] = df_filtered_dong['월'].apply(define_season)

# 결과 확인
print(df_filtered_dong[['영업자시작일', '폐업일자', '총영업일', '월', '계절']])
```
