```python
# 가장 최신의 기준년도와 분기 확인
latest_year = rent_data['STRD_YR_CD'].max()
latest_quarter = rent_data[rent_data['STRD_YR_CD'] == latest_year]['STRD_QTR_CD'].max()

# 최신 데이터 필터링
latest_data = rent_data[(rent_data['STRD_YR_CD'] == latest_year) & (rent_data['STRD_QTR_CD'] == latest_quarter)]

latest_data
```

- 임대료 값이 0으로 누락되어있는 것들이 있음 -> 층수를 고려한 평균으로 대체
- 층별 가중치 고려 : 전체를 1로 보았을 때 1층 - 0.5, 1층외 - 0.3, 전체층 - 0.2 로 부여

```python
# 층별 가중치 고려
floor_weights = {
    '1층': 0.5,
    '1층외': 0.3,
    '전체층': 0.2
}
```

```python
# 층별로 0이 아닌 월임대료 평균과 환산임대료 평균의 평균값 계산
floor_mnth_avg = latest_data[latest_data['MNTH_RENTCG_AVE'] > 0].groupby('LET_CURPRC_FLR_CLSF_CD_NM')['MNTH_RENTCG_AVE'].mean()
floor_exche_avg = latest_data[latest_data['EXCHE_RENTCG_AVE'] > 0].groupby('LET_CURPRC_FLR_CLSF_CD_NM')['EXCHE_RENTCG_AVE'].mean()

# 0 값을 해당 층의 평균값으로 대체
for floor in floor_weights.keys():
    floor_mnth_mean = floor_mnth_avg.get(floor, 0)
    floor_exche_mean = floor_exche_avg.get(floor, 0)

    floor_indices = latest_data[(latest_data['LET_CURPRC_FLR_CLSF_CD_NM'] == floor) & (latest_data['MNTH_RENTCG_AVE'] == 0)].index
    latest_data.loc[floor_indices, 'MNTH_RENTCG_AVE'] = floor_mnth_mean

    floor_indices = latest_data[(latest_data['LET_CURPRC_FLR_CLSF_CD_NM'] == floor) & (latest_data['EXCHE_RENTCG_AVE'] == 0)].index
    latest_data.loc[floor_indices, 'EXCHE_RENTCG_AVE'] = floor_exche_mean


# 변경 후 데이터 확인
latest_data
```

```python
# 가중치를 데이터에 매핑
latest_data['Weight'] = latest_data['LET_CURPRC_FLR_CLSF_CD_NM'].map(floor_weights)

# 가중치가 적용된 평균 계산
latest_data['Weighted_MNTH_RENTCG_AVE'] = latest_data['MNTH_RENTCG_AVE'] * latest_data['Weight']
latest_data['Weighted_EXCHE_RENTCG_AVE'] = latest_data['EXCHE_RENTCG_AVE'] * latest_data['Weight']

# 가중치가 적용된 총합을 행정동별로 그룹화하여 계산
weighted_average = latest_data.groupby('ADSTRD_CD_NM').agg({
    'Weighted_MNTH_RENTCG_AVE': 'sum',
    'Weighted_EXCHE_RENTCG_AVE': 'sum'
}).reset_index()

# 가중치가 적용된 데이터셋
weighted_average
```

### 스케일링
- 임대료 점수 생성

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
standard_scaler = StandardScaler()

# 가중치가 적용된 평균에 대해 표준화 적용
weighted_average[['MNTH_RENTCG_AVE_Standardized', 'EXCHE_RENTCG_AVE_Standardized']] = standard_scaler.fit_transform(
    weighted_average[['Weighted_MNTH_RENTCG_AVE', 'Weighted_EXCHE_RENTCG_AVE']]
)

weighted_average[['ADSTRD_CD_NM', 'MNTH_RENTCG_AVE_Standardized', 'EXCHE_RENTCG_AVE_Standardized']].head()
```

```python
rent_data_std = weighted_average[['ADSTRD_CD_NM', 'MNTH_RENTCG_AVE_Standardized', 'EXCHE_RENTCG_AVE_Standardized']]
rent_data_std
```

```python
# 임대료 상위 5개 지역 찾기
## MNTH_RENTCG_AVE_Standardized = 월 임대료 평균 기준 상위 지역
## EXCHE_RENTCG_AVE_Standardized = 환산 임대료 평균 기준 상위 지역
top_rent_monthly = rent_data_std.sort_values(by='MNTH_RENTCG_AVE_Standardized', ascending=False).head()
top_rent_exche = rent_data_std.sort_values(by='EXCHE_RENTCG_AVE_Standardized', ascending=False).head()

top_rent_monthly, top_rent_exche
```
