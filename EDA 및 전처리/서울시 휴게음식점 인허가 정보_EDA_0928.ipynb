## 서울시 휴게 음식점 정보
https://data.seoul.go.kr/dataList/OA-16095/S/1/datasetView.do

```python
seoul = '/content/drive/MyDrive/baf_society/서울시 휴게음식점 인허가 정보.csv'
seoul_data = pd.read_csv(seoul, encoding='utf-8')
```

```python
seoul_data.info()
```

```python
# 133064개 행 중 업태 구분 명이 커피숍인 데이터만 추출 -> 36521개 커피숍 존재
seoul_cafe = seoul_data[seoul_data['업태구분명'] == '커피숍']
# seoul_cafe
```

- 필요한 변수만을 선택

```python
seoul_cafe = seoul_cafe[['사업장명','인허가일자','상세영업상태코드','상세영업상태명','폐업일자','소재지면적','지번주소','도로명주소','좌표정보(X)','좌표정보(Y)']]
```

```python
seoul_cafe
```

```python
seoul_cafe.head(20)
```

```python
# '메머드', '컴포즈', '메가' 등의 프렌차이즈 카페명을 포함한 행 필터링
franchises = ['메머드', '매머드', '컴포즈', '메가', '빽다방', '더벤티', '이디야', '스타벅스', '커피빈', '할리스', '투썸플레이스','커피빈','더커피빈','카페베네','탐앤탐스','풀바셋','파스쿠찌']
pattern = '|'.join(franchises)
seoul_cafe_filtered = seoul_cafe[~seoul_cafe['사업장명'].str.contains(pattern, regex=True, case=False)]
```

```python
seoul_cafe_filtered.head()
```

```python
seoul_cafe_filtered.info()
```
