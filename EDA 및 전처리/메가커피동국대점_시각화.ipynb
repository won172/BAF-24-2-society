```python
# 선택한 메가엠지씨커피 동국대점 좌표
selected_mega_coords = (37.561343,126.998598)

# 반경 설정 (미터)
radius = 500

# 선택한 메가엠지씨커피 동국대점 주변의 지하철 역 개수
nearby_subway_count_mega = sum(subway_data.apply(lambda row: calculate_distance(row['y'], row['x'], selected_mega_coords[0], selected_mega_coords[1]) <= radius, axis=1))

# 선택한 메가엠지씨커피 동국대점 주변의 버스 정류장 개수
nearby_bus_count_mega = sum(bus_data.apply(lambda row: calculate_distance(row['y'], row['x'], selected_mega_coords[0], selected_mega_coords[1]) <= radius, axis=1))

# 지도 생성
map_mega = folium.Map(location=selected_mega_coords, zoom_start=16)

# 메가엠지씨커피 동국대점 마커 추가
folium.Marker(
    location=selected_mega_coords,
    popup=folium.Popup("메가엠지씨커피 동국대점", max_width=200),
    icon=folium.Icon(color='darkred'),
    tooltip="메가엠지씨커피 동국대점"
).add_to(map_mega)

# 주변 지하철 역 마커 추가
for idx, station in subway_data.iterrows():
    if calculate_distance(station['y'], station['x'], selected_mega_coords[0], selected_mega_coords[1]) <= radius:
        folium.Marker(
            location=[station['y'], station['x']],
            icon=folium.Icon(color='blue', icon='subway', prefix='fa'),
            tooltip=station['역명']
        ).add_to(map_mega)

# 주변 버스 정류장 마커 추가
for idx, bus in bus_data.iterrows():
    if calculate_distance(bus['y'], bus['x'], selected_mega_coords[0], selected_mega_coords[1]) <= radius:
        folium.Marker(
            location=[bus['y'], bus['x']],
            icon=folium.Icon(color='green', icon='bus', prefix='fa'),
            tooltip=bus['정류장명']
        ).add_to(map_mega)

map_mega
```
