## 카페 필터링
- 업태명이 커피숍, 다방, 까페이거나 기타 분류에 있는 것 중 업소명이 카페인 행 추출
- 78722개

```python
cafe_list = ['컴포즈커피', '스타벅스', '투썸플레이스', '이디야', '할리스', '탐앤탐스', '매머드',
             '메가.*커피|메가엠지씨|MGC커피', '빽다방', '더벤티', '커피빈', '디저트39', '셀렉토커피', '커피베이', '바나프레소',
             '파스쿠찌', '블루보틀', '카페베네', '폴바셋', '커피에반하다', '엔제리너스', '테라로사', '감성커피', '하삼동커피','카페','커피']

# cafe_list를 정규 표현식으로 변환합니다.
regex_pattern = '|'.join(cafe_list)  # 모든 카페 이름을 |로 연결합니다.

# 필터링 조건 설정 및 적용
seoul_filtered_cafe = seoul_filtered[
    (seoul_filtered['업태명'].isin(['커피숍', '다방', '까페'])) |
    ((seoul_filtered['업태명'].isin(['기타', '기타 휴게음식점'])) & (seoul_filtered['업소명'].str.contains(regex_pattern, regex=True, case=False)))
]
```

### 카페 데이터 전처리

```python
# '영업자시작일' 열의 날짜 형식유효한지 확인, 아닌것들 결측치 대체
df['영업자시작일_parsed'] = pd.to_datetime(df['영업자시작일'], errors='coerce', format='%Y%m%d')
invalid_dates = df[df['영업자시작일_parsed'].isna()]['영업자시작일']

# 날짜 형식이 아닌것 출력
print("날짜 형식이 아닌 값들:")
print(invalid_dates)
```

```python
# 영업시작일의 결측값을 허가신고일로 대체 -> 결측치 확인되면
# df['영업자시작일'] = df['영업자시작일'].fillna(df['허가신고일'])
# franchises_cafe.info()
```

```python
df = df[['업소명','행정동명','소재지도로명','소재지지번','영업자시작일','소재지시작일','폐업일자','영업장면적(㎡)','업소위치','영업장조리장면적(㎡)','영업장객실면적(㎡)','영업장무도장면적(㎡)','영업장기타면적(㎡)','업소내화장실면적(㎡)','타업소공동화장실면적(㎡)','영업장탈의실면적(㎡)','영업장객석면적(㎡)','작업장면적(㎡)','검사실면적(㎡)']]
```

```python
## 결측치를 0으로 채운뒤 면적 변수끼리의 합을 이용해 총 면적 계산
df['총면적'] = (
    df['영업장면적(㎡)'].fillna(0) +
    df['영업장조리장면적(㎡)'].fillna(0) +
    df['영업장객실면적(㎡)'].fillna(0) +
    df['영업장무도장면적(㎡)'].fillna(0) +
    df['영업장기타면적(㎡)'].fillna(0) +
    df['업소내화장실면적(㎡)'].fillna(0) +
    df['타업소공동화장실면적(㎡)'].fillna(0) +
    df['영업장탈의실면적(㎡)'].fillna(0) +
    df['영업장객석면적(㎡)'].fillna(0) +
    df['작업장면적(㎡)'].fillna(0) +
    df['검사실면적(㎡)'].fillna(0)
)
```
