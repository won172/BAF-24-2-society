```python
# '카페명' 열을 '업소명'으로 초기화합니다.
franchises_cafe_filtered['카페명'] = franchises_cafe_filtered['업소명']

# 각 카페 브랜드에 따라 '카페명' 열을 업데이트합니다.
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains(r'메가.*커피|메가엠지씨', regex=True), '카페명'] = '메가커피'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('컴포즈'), '카페명'] = '컴포즈커피'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('빽다방'), '카페명'] = '빽다방'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('더벤티'), '카페명'] = '더벤티'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('매머드|메머드'), '카페명'] = '매머드'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('스타벅스'), '카페명'] = '스타벅스'
# franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('폴바셋'), '카페명'] = '폴바셋'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('투썸'), '카페명'] = '투썸플레이스'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('할리스'), '카페명'] = '할리스'
# franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('파스쿠찌'), '카페명'] = '파스쿠찌'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('이디야'), '카페명'] = '이디야커피'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('커피빈|더커피빈'), '카페명'] = '커피빈'
# franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('카페베네'), '카페명'] = '카페베네'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('탐앤탐스'), '카페명'] = '탐앤탐스'
# franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('블루보틀'), '카페명'] = '블루보틀'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('디저트39|디저트 39'), '카페명'] = '디저트39'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('셀렉토'), '카페명'] = '셀렉토커피'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('커피베이'), '카페명'] = '커피베이'
franchises_cafe_filtered.loc[franchises_cafe_filtered['업소명'].str.contains('바나프레소'), '카페명'] = '바나프레소'


# 변경된 데이터를 출력하여 확인
print(franchises_cafe_filtered[['업소명', '카페명']].tail(30))
```
